```python
import seaborn as sns
plt.figure(figsize=(12,12))
sns.heatmap(data=df.corr().round(2),annot=True,cmap='coolwarm')
```

```python
df=df[['rm','lstat','ptratio','tax','medv']]
df
```

```python
sns.pairplot(df)
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,3))
plt.subplot(1,2,1)
sns.boxplot(x='medv',data=df,color='blue')
plt.subplot(1,2,2)
sns.distplot(x=df.medv,color='red')
plt.show()
```

```python
medv_q3 = df['medv'].quantile(0.75)
medv_q1 = df['medv'].quantile(0.25)
print(medv_q3, medv_q1)
medv_iqr = medv_q3 - medv_q1
print(medv_iqr)
```

```python
print("shape of dataset before removing outliers:",df.shape)
df1=df[~(df['medv']==50)]
print("shape of dataset after removing outliers:",df1.shape)
```

```python
plt.figure(figsize=(20,3))
plt.subplot(1,3,1)
sns.boxplot(df1.tax,color='blue')
plt.title("boxplot of tax")
plt.subplot(1,3,2)
sns.distplot(a=df1.tax,color='red')
plt.title("Distributio plot")
plt.subplot(1,3,3)
sns.scatterplot(x=df1.tax,y=df1.medv,color='pink')
plt.title("Scatterplot of tax")
plt.show()
```
